**1)What is the average price of food across all restaurants?**

```python
np.round(df['Price'].mean(),2)
```

**2)Which restaurant has the highest average rating?**

```python
df.groupby("Restaurant")["Avg ratings"].mean().sort_values(ascending=False).head(1)
```

**3)Which restaurant has the lowest average rating?**

```python
df.groupby("Restaurant")["Avg ratings"].mean().sort_values().head(1)
```

**4)How does average rating correlate with the price?**

```python
plt.figure(figsize=(11,3))
sns.scatterplot(x=df["Avg ratings"],y=df["Price"],data=df)
plt.show()
```

**5)Which area has the most number of restaurants?**

```python
df.groupby("Area")["Restaurant"].count().sort_values(ascending=False).head(1)
```

**6)Which city has the most number of restaurants?**

```python
df.groupby("City")["Restaurant"].count().sort_values(ascending=False).head(1)
```

**7)What is the average delivery time across all orders?**

```python
df.head(1)
```

```python
np.round(df["Delivery time"].mean(),2)
```

**8)Which food type is the most popular based on total ratings?**

```python
df1=df[["Total ratings","Food type"]]
df1["Food type"]=df1["Food type"].str.split(",")
df1.explode("Food type").groupby("Food type")["Total ratings"].sum().sort_values(ascending=False).head(1)
```

```python
df2=df[["Total ratings","Food type"]]
df2["Food type"]=df2["Food type"].str.split(",")
df3=df2.explode("Food type")
df3.pivot_table(index="Food type",values="Total ratings",aggfunc="sum").sort_values("Total ratings",ascending=False).head(1)
```

**9)Are restaurants with higher prices likely to have higher ratings?**

```python
plt.figure(figsize=(11,3))
sns.scatterplot(x=df["Avg ratings"],y=df["Price"],data=df)
plt.show()
```

```python
df3=df[["Price","Avg ratings","Total ratings","Delivery time"]]
corr=df3.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

**10)Which restaurant has the slowest average delivery time?**

```python
df.groupby("Restaurant")["Delivery time"].mean().sort_values().head(1)
```

**11)Are restaurants with faster delivery times more highly rated?**

```python
df4=df[["Delivery time","Total ratings"]]
corr=df4.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

**12)How does the number of food types offered correlate with average ratings?**

```python
def coun(num):
    val=0
    for i in num:
        val=val+1
    return val

df4=df[["Food type"]]
df4["Food type"]=df4["Food type"].str.split(",")
df4["count"]=df4["Food type"].apply(coun)
df["count1"]=df4["count"]
#a=df[["Avg ratings","count1"]]
#sns.heatmap(a,annot=True,cmap="coolwarm")
#sns.scatterplot(x=df["Avg ratings"],y=df["count1"],data=df)
correlation = df['count1'].corr(df['Avg ratings'])
print(correlation)
 
```

**13)Which restaurant offers the most food types?**

```python
a=df["count1"].max()
b=df[df["count1"]==a]
print(b)
```

**14)Which area in Bangalore has the highest average restaurant rating?**

```python
a=df6["count1"].max()
b=df6[df6["count1"]==a]
print(b)
```
